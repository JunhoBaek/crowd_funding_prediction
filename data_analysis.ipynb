# 데이터 시각화

```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.head(10)
```

```python
df[df['main_category'] == 'design']['sub_category'].value_counts()
```

```python
a = df[df['status'] == 'successful'].groupby('main_category')['status'].count()
b = df[df['status'] == 'failed'].groupby('main_category')['status'].count()
c = a/(a+b)
c = pd.DataFrame(c)
c.head()
```

```python
df.groupby(['main_category', 'status'])['status'].count()
```

## main category별 크라우드 펀딩 성공률

```python
sns.barplot(data = c, x = c.index, y = 'status')
```

```python
comic_df = df[df['main_category'] == 'comics']
comic_df.head()
```

```python
df[df['main_category'] == 'comics']['sub_category'].value_counts()
```

```python
a = comic_df[comic_df['status'] == 'successful'].groupby('sub_category')['status'].count()
b = comic_df[comic_df['status'] == 'failed'].groupby('sub_category')['status'].count()
c = a/(a+b)
c = pd.DataFrame(c)
c.drop(['Comics', 'Comic Books'], axis = 0, inplace = True)
```

## Comics 세부 카테고리별 펀딩 성공률

```python
sns.barplot(data = c, x = c.index, y = 'status')
```
